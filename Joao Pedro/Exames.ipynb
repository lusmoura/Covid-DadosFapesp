```python
import pandas as pd
import numpy as np
import os
```

## Lendo datasets

```python
einstein = pd.read_csv('einstein/einstein_small_dataset_exames.csv', sep='|', engine='python')
einstein = einstein[['id_paciente', 'dt_coleta', 'de_origem', 'de_exame', 'de_analito', 
                            'de_resultado', 'cd_unidade', 'de_valor_referencia']].copy()

rename_cols = {'id_paciente':'ID_PACIENTE',
               'dt_coleta':'DT_COLETA',
               'de_origem':'DE_ORIGEM',
               'de_exame':'DE_EXAME',
               'de_analito':'DE_ANALITO', 
               'de_resultado':'DE_RESULTADO',
               'cd_unidade':'CD_UNIDADE',
               'de_valor_referencia':'DE_VALOR_REFERENCIA'}

einstein = einstein.rename(columns=rename_cols)
```

```python
hsl = pd.read_csv('hsl/small/COVID_LAB_RESULT.csv', sep='|')
hsl = hsl[['ID_PACIENTE', 'DT_COLETA', 'DE_ORIGEM', 'DE_EXAME', 
                 'DE_ANALITO', 'DE_RESULTADO', 'CD_UNIDADE', 'DE_VALOR_REFERENCIA']].copy()
```

```python
fleury = pd.read_csv('fleury/Grupo_Fleury_Small_Dataset_Covid19_Resultados_Exames_1.csv', sep='|', encoding = "ISO-8859-1")
fleury = fleury[['ID_PACIENTE', 'DT_COLETA', 'DE_ORIGEM', 'DE_EXAME', 
                       'DE_ANALITO', 'DE_RESULTADO', 'CD_UNIDADE', 'DE_VALOR_REFERENCIA']].copy()
```

## Analitos

```python
fleury_analito = set(fleury['DE_ANALITO'].unique())
print(f"Número de analitos Fluery: {len(fleury_analito)}")
```

```python
hsl_analito = set(hsl['DE_ANALITO'].unique())
print(f"Número de analitos Sírio Libanês: {len(hsl_analito)}")
```

```python
einstein_analito = set(einstein['DE_ANALITO'].unique())
print(f"Número de analitos Einstein: {len(einstein_analito)}")
```

```python
intersection_hsl_fleury = fleury_analito.intersection(hsl_analito)
print(f"Número de analitos em comum entre Fleury e HSL: {len(intersection_hsl_fleury)}")

intersection_einstein_fleury = fleury_analito.intersection(einstein_analito)
print(f"Número de analitos em comum entre Fleury e Einstein: {len(intersection_einstein_fleury)}")

intersection_einstein_hsl = hsl_analito.intersection(einstein_analito)
print(f"Número de analitos em comum entre HSL e Einstein: {len(intersection_einstein_hsl)}")

intersection_total = intersection_hsl_fleury.intersection(einstein_analito)
print(f"Número de analitos em comum entre todos: {len(intersection_total)}")
```

## Exame

```python
fleury_exame = set(fleury['DE_EXAME'].unique())
print(f"Número de exames Fluery: {len(fleury_exame)}")
```

```python
hsl_exame = set(hsl['DE_EXAME'].unique())
print(f"Número de exames Sírio Libanês: {len(hsl_exame)}")
```

```python
einstein_exame = set(einstein['DE_EXAME'].unique())
print(f"Número de exames Einstein: {len(einstein_exame)}")
```

```python
intersection_hsl_fleury_exames = fleury_exame.intersection(hsl_exame)
print(f"Número de exames em comum entre Fleury e HSL: {len(intersection_hsl_fleury_exames)}")

intersection_einstein_fleury_exames = fleury_exame.intersection(einstein_exame)
print(f"Número de exames em comum entre Fleury e Einstein: {len(intersection_einstein_fleury_exames)}")

intersection_einstein_hsl_exames = hsl_exame.intersection(einstein_exame)
print(f"Número de exames em comum entre HSL e Einstein: {len(intersection_einstein_hsl_exames)}")

intersection_total_exames = intersection_hsl_fleury_exames.intersection(einstein_exame)
print(f"Número de exames em comum entre todos: {len(intersection_total_exames)}")
```

## Concatenando datasets

```python
full = pd.concat([hsl, fleury, einstein])
full['ID_PACIENTE'] = full['ID_PACIENTE'].astype(str)
```
