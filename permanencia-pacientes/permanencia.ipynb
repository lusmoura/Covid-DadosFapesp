```python
import pandas as pd
import numpy as np
from datetime import timedelta

import seaborn as sns
import matplotlib.pyplot as plt
```

## Hospital Sírio Libanês

```python
hsl = pd.read_csv('../Dados/SirioLibanes/data/hsl_lab_result_1.csv', sep='|')
hsl = hsl[['ID_PACIENTE', 'DT_COLETA', 'DE_ORIGEM', 'DE_EXAME', 
                 'DE_ANALITO', 'DE_RESULTADO', 'CD_UNIDADE', 'DE_VALOR_REFERENCIA']].copy()
```

```python
hsl['DT_COLETA_MAX'] = hsl['DT_COLETA'].copy()
hsl = hsl.rename(columns={'DT_COLETA':'DT_COLETA_MIN'})

hsl['DT_COLETA_MIN'] = pd.to_datetime(hsl['DT_COLETA_MIN'])
hsl['DT_COLETA_MAX'] = pd.to_datetime(hsl['DT_COLETA_MAX'])

permanencia = hsl.groupby('ID_PACIENTE').agg({'DT_COLETA_MIN':'min', 'DT_COLETA_MAX':'max'})

permanencia['Tempo'] = permanencia['DT_COLETA_MAX'] - permanencia['DT_COLETA_MIN'] + timedelta(days=1)
permanencia['Tempo'] = permanencia['Tempo'].dt.days

permanencia.head()
```

```python
plt.figure(figsize=(16,9))
plt.title(f'Tempo de permanência - Hospital Sírio Libanês | {permanencia.shape[0]} pacientes')
sns.distplot(permanencia['Tempo'], bins=permanencia['Tempo'].max()//7);
```

```python
plt.figure(figsize=(16,9))
plt.title(f'Tempo de permanência - Hospital Sírio Libanês (> 7 dias) | {permanencia[permanencia["Tempo"] > 7].shape[0]} pacientes')
sns.distplot(permanencia[permanencia['Tempo'] > 7]['Tempo'], bins=permanencia['Tempo'].max()//7 - 1);
```
