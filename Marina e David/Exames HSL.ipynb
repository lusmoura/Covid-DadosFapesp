## Processando

```python
hsl['DE_VALOR_REFERENCIA'] = hsl['DE_VALOR_REFERENCIA'].replace(np.nan, 'NaN')
hsl['CD_UNIDADE'] = hsl['CD_UNIDADE'].replace(np.nan, 'NaN') 
```

```python
hsl.loc[hsl['DE_VALOR_REFERENCIA'].str.contains('\d+', regex=True), 'DE_RESULTADO'] = 'Number'
```

```python
hsl.head()
```

```python
def check_float(string):
    try:
        number = float(string)
        return 'Number'
    except:
        return string
```

```python
hsl['DE_RESULTADO'] = hsl['DE_RESULTADO'].str.replace(',', '.')
hsl['DE_RESULTADO'] = hsl['DE_RESULTADO'].apply(lambda x: check_float(x))
```
